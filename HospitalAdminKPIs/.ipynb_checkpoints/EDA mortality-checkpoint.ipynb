```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mortality_data = pd.read_csv("./data/archive/HDHI Mortality Data.csv")
    
mortality_data.head()
                             
```

```python
print(mortality_data.info())

pd.options.display.float_format = '{:.2f}'.format
print(mortality_data.describe())
print(mortality_data.isnull().sum())
```

```python
mrd_mean = mortality_data["MRD"].mean()
mortality_data["MRD"].fillna(mrd_mean, inplace=True)
print(mortality_data)
print(mortality_data.isnull().sum())
```

```python
plt.figure(figsize=(8,6))
sns.histplot(mortality_data['AGE'], bins =30, kde = True, color = 'Red')
plt.title ("Mortality Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
mortality_data['DATE OF BROUGHT DEAD'] = pd.to_datetime(mortality_data ['DATE OF BROUGHT DEAD'], format = '%m%d%y')
mortality_data['YEAR OF DEATH'] = mortality_data['DATE OF BROUGHT DEAD'].dt.year
deaths_per_year = mortality_data['YEAR OF DEATH'].value_counts().sort_index()

plt.figure(figsize = (8,6))
deaths_per_year.plot(kind = 'bar' , color = 'blue')
plt.title (" Year of deaths")
plt.xlabel ("Year")
plt.ylabel("Count")
plt.show()

yearly_counts = mortality_data['YEAR OF DEATH'].value_counts()
print(" Count per year: ")
print(yearly_counts)                            
```

```python
rural_count = (mortality_data['RURAL/URBAN'] == "R").sum()
urban_count = (mortality_data['RURAL/URBAN'] == "U").sum()

total = len(mortality_data)
rural_percentage= (rural_count / total) * 100
urban_percentage= (urban_count / total) * 100

labels = "Rural", "Urban"  
sizes = [rural_percentage , urban_percentage]
explode = (0,0.1)

fig1, ax1 = plt.subplots()
ax1.pie (sizes, explode = explode , labels = labels , autopct = '%1.1f%%' , shadow = True , startangle = 90)
ax1.axis =( 'equal')
plt.show

rural_counts = (mortality_data['RURAL/URBAN'] == 'R').sum()
urban_counts = (mortality_data['RURAL/URBAN'] == 'U').sum()
print(f'RURAL COUNT : {rural_counts} , URBAN COUNT: {urban_counts}')
```

```python
# calculates frequency of each age group
age_groups = mortality_data['AGE'].value_counts().sort_index()


plt.figure(figsize=(25, 10))
plt.bar(age_groups.index, age_groups.values, color='skyblue')
plt.xlabel('Age Groups')
plt.ylabel('Frequency')
plt.title('Frequency of Age Groups')
plt.grid(True)
plt.xticks(age_groups.index)  # Sets x-axis ticks to be the age values
plt.show()
```

```python
male_count =(mortality_data['GENDER '] == 'M').sum()
female_count =(mortality_data['GENDER '] == 'F').sum()
print(f"Males : {male_count} , Females : {female_count}")
      
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(mortality_data['AGE'], mortality_data['MRD'], color='green', alpha=0.6)
plt.title('MRD vs Age')
plt.xlabel('Age')
plt.ylabel('MRD')
plt.grid(True)
plt.ticklabel_format(style = 'plain')
plt.tight_layout()
plt.show()
```

```python
# Separate data by gender
male_data = mortality_data[mortality_data['GENDER '] == 'M']
female_data = mortality_data[mortality_data['GENDER '] == 'F']


plt.figure(figsize=(10, 6))
plt.hist(male_data['AGE'], bins=10, alpha=0.6, color='blue', label='Male')
plt.hist(female_data['AGE'], bins=10, alpha=0.6, color='pink', label='Female')

plt.title('Age Distribution by Gender')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```
