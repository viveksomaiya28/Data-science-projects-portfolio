```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pollution_data = pd.read_csv("./data/archive/HDHI Pollution Data.csv")
pollution_data.head()
```

```python

pollution_data['DATE'] = pd.to_datetime(pollution_data['DATE'])

# List of columns that should be numeric
numeric_columns = ['AQI', 'PM2.5 AVG', 'PM2.5 MIN', 'PM2.5 MAX', 'PM10 AVG', 'PM10 MIN', 'PM10 MAX',
                   'NO2 AVG', 'NO2 MIN', 'NO2 MAX', 'NH3 AVG', 'NH3 MIN', 'NH3 MAX', 'SO2 AVG', 'SO2 MIN', 'SO2 MAX',
                   'CO AVG', 'CO MIN', 'CO MAX', 'OZONE AVG', 'OZONE MIN', 'OZONE MAX', 'MAX TEMP', 'MIN TEMP', 'HUMIDITY']

# Convert all relevant columns to numeric, forcing errors to NaN
for column in numeric_columns:
    pollution_data[column] = pd.to_numeric(pollution_data[column], errors='coerce')

# Fill missing values with the mean of the respective columns
for column in ['PM10 MAX', 'NO2 MAX', 'OZONE MAX']:
    pollution_data[column].fillna(pollution_data[column].mean(), inplace=True)

# Verify no missing values remain
print(pollution_data.isnull().sum())
```

```python
print(pollution_data.describe())
```

```python
# Correlation between MIN TEMP and HUMIDITY
min_temp_humidity_corr = pollution_data['MIN TEMP'].corr(pollution_data['HUMIDITY'])
print(f'Correlation between MIN TEMP and HUMIDITY: {min_temp_humidity_corr}')

# Correlation between MAX TEMP and HUMIDITY
max_temp_humidity_corr = pollution_data['MAX TEMP'].corr(pollution_data['HUMIDITY'])
print(f'Correlation between MAX TEMP and HUMIDITY: {max_temp_humidity_corr}')
```

```python
# Convert DATE column to datetime
pollution_data['DATE'] = pd.to_datetime(pollution_data['DATE'])

# Extract the year from the DATE column
pollution_data['YEAR'] = pollution_data['DATE'].dt.year

# Plot histogram of AQI values by year
plt.figure(figsize=(10, 6))
for year in pollution_data['YEAR'].unique():
    plt.hist(pollution_data[pollution_data['YEAR'] == year]['AQI'], bins=20, alpha=0.5, label=str(year))

# Adding labels and title
plt.xlabel('AQI')
plt.ylabel('Frequency')
plt.title('Histogram of AQI Values by Year')
plt.legend(title='Year')
plt.tight_layout()
plt.show()

```

```python


# Convert DATE column to datetime
pollution_data['DATE'] = pd.to_datetime(pollution_data['DATE'])

# List of columns that should be numeric
numeric_columns = ['AQI', 'PM2.5 AVG', 'PM2.5 MIN', 'PM2.5 MAX', 'PM10 AVG', 'PM10 MIN', 'PM10 MAX',
                   'NO2 AVG', 'NO2 MIN', 'NO2 MAX', 'NH3 AVG', 'NH3 MIN', 'NH3 MAX', 'SO2 AVG', 'SO2 MIN', 'SO2 MAX',
                   'CO AVG', 'CO MIN', 'CO MAX', 'OZONE AVG', 'OZONE MIN', 'OZONE MAX', 'MAX TEMP', 'MIN TEMP', 'HUMIDITY']

# Convert all relevant columns to numeric, forcing errors to NaN
for column in numeric_columns:
    pollution_data[column] = pd.to_numeric(pollution_data[column], errors='coerce')

# Check for non-numeric values
non_numeric_counts = pollution_data[numeric_columns].apply(lambda x: x.isna().sum())
print(non_numeric_counts)

# Calculate the correlation matrix
correlation_matrix = pollution_data[numeric_columns].corr()

# Plot the heatmap
plt.figure(figsize=(18, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Plot PM10 AVG vs AQI
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PM10 AVG', y='AQI', data=pollution_data, alpha=0.7)
plt.title('Scatter Plot of PM10 AVG vs AQI')
plt.xlabel('PM10 AVG')
plt.ylabel('AQI')
plt.grid(True)
plt.show()
```

```python
# Scatter plot for PM2.5 AVG vs AQI
plt.scatter(pollution_data['PM2.5 AVG'], pollution_data['AQI'], color='blue', label='PM2.5 AVG', alpha=0.7)

# Scatter plot for PM10 AVG vs AQI
plt.scatter(pollution_data['PM10 AVG'], pollution_data['AQI'], color='green', label='PM10 AVG', alpha=0.7)


plt.title('Scatter Plot of PM2.5 AVG and PM10 AVG vs AQI')
plt.xlabel('Pollutant Average Levels')
plt.ylabel('AQI')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x ='HUMIDITY', y='AQI', data = pollution_data, alpha= 0.7)
plt.title("Humidty vs AQI")
plt.xlabel('HUMIDITY')
plt.ylabel("AQI")
plt.grid(True)
plt.show()
```
